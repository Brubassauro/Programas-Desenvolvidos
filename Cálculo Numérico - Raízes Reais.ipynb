```
# PROGRAMA DESENVOLVIDO NA DISCIPLINA DE CÁLCULO NUMÉRICO
# O PROGRAMA UTILIZA DIFERENTES MÉTODOS PARA ENCONTRAR RAÍZES IRRACIONAIS DE FUNÇÕES E OS COMPARA.

import numpy as np
import matplotlib.pyplot as plt

#Exercicio a)

def f(x): #Função f(x)
  f=x**3-np.cos(x**2)
  return f

def plot1(x1=-1,x2=2,n=0.01): #Função para grafico de f(x)
  xl=[]
  yl=[]
  for i in range(int((x2-x1)/n)):
    xl=xl+[x1+i*n] #Lista de valores de x
    yl=yl+[f(x1+i*n)] #Lista de valores correspondentes de f(x)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.plot(xl,yl,color='red')
  plt.arrow(x1,0,x2-x1-0.05,0,color='black',head_width=0.2,head_length=0.05)
  plt.arrow(0,min(yl),0,max(yl)-min(yl),color='black',head_width=0.05,head_length=0.2)
  plt.title('f(x) = x³-cos(x²)')
  plt.grid()
  plt.show()
  
def mbissecao(x1,x2,E=1e-16): #Metodo da Bissecao
  xm=(x1+x2)/2
  e=np.abs(f(xm)) #Erro definido como a distancia entre f(x0) e 0
  if f(x1)/np.abs(f(x1))==f(x2)/np.abs(f(x2)): #Checa se o intervalo contem a raiz
    print('Intervalo invalido')
    return
  n=0 #Numero de iteracoes
  while e>=E: #Processo iterativo
    if f(x1)/np.abs(f(x1))==f(xm)/np.abs(f(xm)):
      x1=xm
    else: x2=xm
    xm=(x1+x2)/2
    e=np.abs(f(xm)) #Atualiza o erro
    n+=1 #Contador de iteracoes
  print('a) x0=%s, pelo metodo da Bissecao' %(xm))
  print('Foram feitas %s iteracoes' %(n))
  
plot1()
mbissecao(0,1)

#Exercicio b)

def df(x): #Derivada de f(x)
  df=3*(x**2)+2*x*np.sin(x**2)
  return df
  
def nraphson(x0,E=1e-16): #Metodo de Newton-Raphson
  xn=x0
  e=np.abs(f(xn)) #Erro definido como a distancia entre f(x0) e 0
  n=0 #Numero de iteracoes
  while e>=E: #Processo iterativo
    xn=xn-(f(xn)/df(xn)) #xn+1
    e=np.abs(f(xn)) #Atualiza o erro
    n+=1 #Contador de iteracoes
  print('b) x0=%s, pelo metodo de Newton-Raphson' %(xn))
  print('Foram feitas %s iteracoes' %(n))

nraphson(1)

#Exercicio c) i)

k=14.4
v0=667
r0=0.290

def V(r): #Potencial V(r)
  v=-(k/r)+v0*np.exp(-r/r0)
  return v

def F(r): #Forca F(r)
  f=-(k/(r**2))+v0*np.exp(-r/r0)/r0
  return f

def plot2(x1=0.01,x2=10,n=0.01): #Graficos de V(r) e F(r)
  print('c) i)')
  xl=[]
  yl=[]
  for i in range(int((x2-x1)/n)):
    xl=xl+[x1+i*n]
    yl=yl+[V(x1+i*n)]
  plt.plot(xl,yl,color='green')
  plt.title('V(r) x r')
  plt.grid()
  plt.ylim(-10,10)
  plt.xlabel('r [Å]')
  plt.arrow(-0.5,0,10.7,0,color='black',head_width=0.3,head_length=0.1)
  plt.ylabel('V [eV]')
  plt.show()
  xl=[]
  yl=[]
  for i in range(int((x2-x1)/n)):
    xl=xl+[x1+i*n]
    yl=yl+[F(x1+i*n)]
  plt.plot(xl,yl,color='blue')
  plt.arrow(-0.5,0,10.7,0,color='black',head_width=0.3,head_length=0.1)
  plt.ylim(-5,10)
  plt.title('F(r) x r')
  plt.xlabel('r [Å]')
  plt.ylabel('F [eV/Å]')
  plt.grid()
  plt.show()
  
plot2()

#c) ii)
  
def msecantes(x1,x2,E=1e-16): #Metodo das secantes
  e=np.abs(F(x2)) #Erro definido como a distancia entre F(r0) e 0
  n=0 #Numero de iteracoes
  while e>=E: #Processo iterativo
    xn=x2-((x2-x1)*F(x2))/(F(x2)-F(x1)) #xn+1
    x1=x2
    x2=xn
    n+=1 #Contador de iteracoes
    e=np.abs(F(xn)) #Atualiza o erro
  print('ii) r0=%sÅ, pelo metodo das secantes' %(xn))
  print('Foram feitas %s iteracoes' %(n))

msecantes(1,2)
```
