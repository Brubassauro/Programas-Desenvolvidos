```
# O PROGRAMA USA OS CONCEITOS DE PILHA E NOTAÇÃO PÓSFIXA PARA PRODUZIR UMA INTERAFACE 
# QUE EXECUTA CONTAS E SALVA VARIÁVEIS SEMELHANTE AO TERMINAL DO LINUX

def prioridade(x): # Lista de prioridades
  if x=='+' or x=='-': return 1
  elif x=='*' or x=='/': return 2
  elif x=='^': return 3
  elif x=='(': return 4
  elif x==')': return 5
  elif x=='_' or x=='#': return 6
  else: return 0

class pilha: # Definindo a classe pilha e suas funções
  def __init__(self):
    self.pilha = []
    
  def __len__ (self):
    return len(self.pilha)
  
  def empty(self):
    return len(self.pilha)==0
  
  def push(self, e):
    self.pilha.append(e)

  def top(self):
    if self.empty( ):
      raise Empty("Pilha vazia")
    return self.pilha[-1]

  def pop(self):
    if self.empty( ):
      raise Empty("Pilha vazia")
    return self.pilha.pop( )

def listagem(X): # Transforma uma string em uma lista com os operadores e operandos
  a=''
  for i in range(len(X)):
    if prioridade(X[i])>=1:
      a+=' '+X[i]+' '
    else: a+=X[i]
  a=a.replace('  ',' ')
  a=a.split(' ')
  if a[0]=='': a.pop(0)
  if a[-1]=='':a.pop(-1)
  return a

def substitui(X): # Substitui o nome das variáveis salvas no dicionário pelo valor delas
  for i in range(len(X)):
    if X[i] in Salvo:
      X[i]=Salvo[X[i]]
  return X

def posfixa(x): # Transforma uma string em notação pósfixa em forma de lista. TraduzPosFixa(exp)
  x=x.replace('**','^')
  x=listagem(x)
  if x[0]=='+': x[0]='#'
  if x[0]=='-': x[0]='_'
  for i in range(len(x)-1):
    if x[i+1]=='-':
      if prioridade(x[i])!=0 and prioridade(x[i])!=5:
        x[i+1]='_'
    if x[i+1]=='+':
      if prioridade(x[i])!=0 and prioridade(x[i])!=5:
        x[i+1]='#'
  y=[]
  X=pilha()
  Y=pilha()
  for i in range(len(x)):
    if prioridade(x[i])==0: y+=[x[i]]
    if prioridade(x[i])==1 or prioridade(x[i])==2 or prioridade(x[i])==3 or prioridade(x[i])==6:
      while pilha.__len__(X)!=0:
        if prioridade(pilha.top(X))>=prioridade(x[i]):
          if prioridade(pilha.top(X))==4: break
          y+=[pilha.pop(X)]
        else:
          pilha.push(Y,pilha.pop(X))
      for j in range(pilha.__len__(Y)):
        pilha.push(X,pilha.pop(Y))
      pilha.push(X,x[i])
    if prioridade(x[i])==4: pilha.push(X,x[i])
    if prioridade(x[i])==5:
      for j in range(pilha.__len__(X)):
        if prioridade(pilha.top(X))==4:
          pilha.pop(X)
          break
        y+=[pilha.pop(X)]
  for j in range(pilha.__len__(X)):
    y+=[pilha.pop(X)]
  return y

def calcula(x): # Entra com uma conta em string e retorna o valor da conta. CalcPosFixa(exp)
  X=pilha()
  y=posfixa(x)
  y=substitui(y)
  while len(y)>=1:
    if prioridade(y[0])==0:
      pilha.push(X,y.pop(0))
      continue
    elif y[0]=='+':
      y.pop(0)
      a=float(pilha.pop(X))
      b=float(pilha.pop(X))
      y=[a+b]+y
    elif y[0]=='*':
      y.pop(0)
      a=float(pilha.pop(X))
      b=float(pilha.pop(X))
      y=[a*b]+y
    elif y[0]=='-':
      y.pop(0)
      a=float(pilha.pop(X))
      b=float(pilha.pop(X))
      y=[b-a]+y
    elif y[0]=='/':
      y.pop(0)
      a=float(pilha.pop(X))
      b=float(pilha.pop(X))
      y=[b/a]+y
    elif y[0]=='^':
      y.pop(0)
      a=float(pilha.pop(X))
      b=float(pilha.pop(X))
      y=[b**a]+y
    elif y[0]=='_':
      y.pop(0)
      a=float(pilha.pop(X))
      y=[-a]+y
    elif y[0]=='#':
      y.pop(0)
      a=float(pilha.pop(X))
      y=[+a]+y
  return pilha.top(X)

Salvo={} # Dicionário com as variáveis salvas

def main():
  x=str(input('>>>'))
  x=x.replace(' ','')
  if '=' in x:
    x=x.split('=')
    if x[0] in Salvo:
      Salvo[x[0]]=calcula(x[1])
    else: Salvo.update({x[0]:calcula(x[1])})
  else:
    print(calcula(x))
  main()
      
main()
```
