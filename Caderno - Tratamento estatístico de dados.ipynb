```
import numpy as np
import matplotlib.pyplot as plt
def mod(x):
  y = 0
  if x < 0:
    y += -x
  else:
    y = x
  return y
def metodo_inversao(N=1):
    """
    Gera um vetor de N elementos
    """
    
    # Exemplo para f(x) = 2 * x em [0, 1]    
    inv_g = lambda g : np.sqrt(g)
    
    g = np.random.rand(N)
    x = inv_g(g)
    
    return x
def metodo_exclusao(N=1):
    """
    Gera um vetor de N elementos
    """
    
    # Exemplo para f(x) = 1 - |x| em [0, 1]
    xmin = -1
    xmax = 1
    ymax = 1
    f = lambda x : 1 - mod(x)
    
    i = 0
    x = np.zeros(N)
    while i < N:
        x_cand = xmin + (xmax - xmin) * np.random.rand()
        y_test = ymax * np.random.rand()
        if y_test <= f(x_cand):
            x[i] = x_cand
            i += 1
    
    return x
y = metodo_exclusao(10000)
n = 0
for i in range(len(y)):
  if (np.mean(y) - np.std(y)) <= y[i] <= (np.mean(y) + np.std(y)):
    n+= 1
print("m√©dia:",np.mean(y))
print("Desvio padr√£o da m√©dia:",np.std(y)/np.sqrt(len(y)))
print("desvio padr√£o:",np.std(y))
print(n)
print(n/len(y))
plt.hist(y)
plt.show(y)
```
